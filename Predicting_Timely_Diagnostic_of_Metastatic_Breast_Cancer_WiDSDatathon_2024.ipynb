#### Missing Data

Most of the machine learning models are not able to handle missing data. We therefore need to deal with missing data.

The following are possible solutions to deal with missing data.
1. If the percentage of the missing data is not too high(say <10%), replace the missing data. There are several methods that you can apply to compute the missing data.
2. If the percentage of the missing data is too high drop the variables that are affected.However, droping variable is not very advisable especially a number of them have a high number of missing data. **Always try to understand why the data is missing.**



```python
#check the percentage of the features with the missing values:

# missing_values = training_df.isnull().sum()
# missing_percentage = (missing_values / training_df.shape[0]) * 100
# missing_data = pd.DataFrame({'Missing Values': missing_values, 'Missing Percentage': missing_percentage})
# missing_data = missing_data[missing_data['Missing Values'] > 0]
# missing_data.sort_values(by = 'Missing Percentage',ascending = False)


def var_missing_data(df):
    """return a dataframe with the features that have missing values
        and the percentage that is missing
    """
    missing_values = df.isnull().sum()
    missing_percentage = (missing_values / df.shape[0]) * 100
    missing_data = pd.DataFrame({'Missing Values': missing_values, 'Missing Percentage': missing_percentage})
    missing_data = missing_data[missing_data['Missing Values'] > 0]
    missing_data.sort_values(by = 'Missing Percentage',ascending = False)
    return missing_data

missing_data = var_missing_data(training_df)

missing_data
```

The rest we will impute the missing data points






```python
##drop the features with the high missing values
#drop the variables with more than 50% missing values
# variables_with_missing_data = missing_data[missing_data['Missing Percentage']> 40].index.tolist()


# training_df.drop(variables_with_missing_data, axis=1, inplace=True)


def drop_high_missing_var(df,missing_var_df,per = 40):

  """ Drop the features with very high missing values.
      By default any feature with more than 40% missing values is dropped(this can be adjusted)."""

  vars_to_drop = missing_var_df[missing_var_df['Missing Percentage']> per].index.tolist()


  df_updated = df.drop(vars_to_drop, axis=1)


  return df_updated


training_df =drop_high_missing_var(training_df,missing_data)
```

```python
def replace_missing_var(df):


    """replace the missing values
    Categorical features: Are replaced with the most frequent(Mode)
    Numeric Features are replaced with the Median"""

    #get the missing variables

    df_missing = var_missing_data(df)

    #split the data into categorical and continous

    categorical_vars = df_missing[df.dtypes == 'object']
    continuous_vars = df_missing[df.dtypes != 'object']



    # for the categorical variables replace with the mode
    highest_count_categories = df[categorical_vars.index].mode().iloc[0]
    # Replace missing values with the highest count category
    df[categorical_vars.index] = df[categorical_vars.index].fillna(highest_count_categories)



    #for the continuous variables replace with the median
    mean_continous_variables = df.median().iloc[0]
    #Replace the missing values with the median


    df[continuous_vars.index] = df[continuous_vars.index].fillna(mean_continous_variables)

    return df


updated_training_df = replace_missing_var(training_df)
```

### Compute the correlation matrix and drop the highly correlated features
In this context, we examine the correlation among the features. If there's a strong correlation (either positive or negative, approximately ±0.90) between n features, we discard n-1 of them, retaining only one. It's important to note that there isn't a specific threshold for defining high correlation between variables. Nevertheless, correlated features essentially convey identical information to the model. Consequently, there is no benefit in including the additional n-1 features.



```python
correlation_matrix = continuous_vars.corr()

highly_correlated_features = set()
for i in range(len(correlation_matrix.columns)):

    for j in range(i):


        #print(abs(correlation_matrix.iloc[i, j]))
        if abs(correlation_matrix.iloc[i, j]) > 0.90:
            colname_i = correlation_matrix.columns[i]
            colname_j = correlation_matrix.columns[j]

            # Drop the feature with the highest mean value
            if continuous_vars[colname_i].mean() > continuous_vars[colname_j].mean():
                highly_correlated_features.add(colname_j)
            else:
                highly_correlated_features.add(colname_i)

# Drop highly correlated features
continuous_vars_not_corr = continuous_vars.drop(columns=highly_correlated_features)

len(continuous_vars.columns),len(continuous_vars_not_corr.columns)
```

### **Feature Engineering**

### Continuous Variables
Certain continuous variables indicate the patient's status based on their residential location, expressed as percentages across various categories within a specific variable. For instance, the variable 'RACE' includes sub-variables such as 'race_white', 'race_black', 'race_asian', 'race_native', 'race_pacific', 'race_other', and 'race_multiple'. Typically, patients are more likely to belong to the predominant group within these categories. Thus, we calculate the maximum percentage for each observation among the variables representing 'RACE' and retain the highest value.


We will retain the original features because they help us to understand the social status of the patient based on their residential location


```python
# prompt: get the maximum value per row and return the column header as the entry

import numpy as np

def get_max_value_column(row):
    """
    This function takes a row of a DataFrame and returns the column header of the maximum value.

    Args:
        row (pd.Series): A row of a DataFrame.

    Returns:
        str: The column header of the maximum value.
    """

    max_value = np.max(row)
    max_column = row.idxmax()

    return max_column





def attr_update(subset_col,new_col_name,org_df=updated_training_df):

  subset_df = org_df[subset_col]
  new_col = subset_df.apply(get_max_value_column, axis=1).tolist()
  ##drop the columns that have been combined
  updated_df= org_df.drop(columns = subset_col,axis = 1)
  #add the new column to the data set
  updated_df[new_col_name] = new_col


  return updated_df


```

```python
summ_vars_dummies = pd.get_dummies(summ_training_df[['education','income']], columns=['education','income'], drop_first=True)
```
